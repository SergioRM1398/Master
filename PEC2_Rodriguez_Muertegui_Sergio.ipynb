### Ejercicio 5
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Usando la librera seaborn y el dataset *flights* provedo por la misma librera, crea un grfico de lneas que muestre el nmero total de pasajeros por ao.

```python
flights= sns.load_dataset('flights') #Cargamos el dataset flights de la libreria seaborn
aos= flights.groupby('year')['passengers'].sum() #Agrupamos el total de pasajeros por aos

fig, ax= plt.subplots() #Creamos figura

ax.plot(aos.index.tolist(),aos.values,
        marker='o',
        markersize= 4 ) #Realizamos el plot con los ejes correspondientes

ax.set_xlabel('Ao') #Colocamos el ttulo y los nombres de los ejes
ax.set_ylabel('Nmero de pasajeros')
ax.set_title('Nmero de pasajeros por ao')

plt.show() #Lo mostramos 
```

### Ejercicio 3
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    Utilizando el siguiente dataset con informacin de pinginos, y asumiendo que cada fila representa un pingino, calcula el nmero de pinginos por especie e isla y ordena los resultados de forma descendente. A continuacin muestra el resultado en un pie chart creado con matplotlib (si hay, por ejemplo, 5 combinaciones de especie e isla, el grfico deber tener 5 sectores de crculo). Aade a cada sector circular una etiqueta que indique la especie e isla que representa.
</div>

```python
penguins=sns.load_dataset('penguins') #Cargamos le dataframe tips de la libreria Seaborn

cantidad=penguins.groupby(['species', 'island']).size() #Agrupamos la cantidad de pinguinos (funcin .size()) segun la especie y la isla
cantidad=cantidad.sort_values(ascending=False) #Los ordenamos de forma descendente
print(cantidad) #Lo mostramos

fig, ax= plt.subplots() #Creamos la figura
ax.pie(x= cantidad.values,
       labels= cantidad.index.tolist())
#Creamos el pie chart con las etiquetas correspondientes
ax.set_title('Pinguinos por Especie e Isla') #Ponemos un ttulos

plt.show()
```

### Ejercicio 2
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Utilizando el dataset titanic de Seaborn, calcula la tasa de supervivencia (el porcentaje de personas que sobrevivieron) por clase y sexo. Representa estos resultados en un grfico de barras agrupadas, donde el eje x represente las clases y los grupos de barras representen el sexo, tal y como se muestra en la imagen.
</div>

```python
titanic= sns.load_dataset('titanic') #Cargamos el dataset de planets
titanic.head()
sup= titanic.groupby(['pclass', 'sex'])['survived'].sum() #Agrupamos usando sum() ya que los que sobreviven son y los que no, son 0, no aportan 

todos= titanic.groupby(['pclass', 'sex'])['survived'].size() #Agrupamos con size() para obtener supervivientes y no supervivientes

porcsuperc=(sup.values/todos.values)*100 #Calculamos el porcentaje con los valores de las agrupaciones

porcentaje= pd.Series(data=porcsuperc, index= sup.index ) #Creamos una nueva serie con los porcentajes
print(porcentaje) #Los mostramos en pantalla

porcentaje=porcentaje.reset_index() #Transformamos la serie en un dataframe para poderlo tranajar comodamente
porcentaje=porcentaje.rename(columns={0:'Porcentaje', 'pclass': 'Clase', 'sex': 'Sexo'}) #Cambiamos los nombres de las columnas
porcentaje['Clase']= porcentaje['Clase'].replace({1:'First', 2:'Second', 3:'Third'}) #Y los valores de la clase para que se asemeje ms al resultado

fig, ax= plt.subplots(figsize=(10,6)) #Creamos la figura
sns.barplot(data=porcentaje, x='Clase', y='Porcentaje', hue= 'Sexo') #Usamos la funcin barplot()
ax.set_ylabel('Tasa de Supervivencia (%)') #Ponemos los nombres de los ejes
ax.set_title('Tasa de Supervivencia por Clase y Sexo en el Titanic') #Ponemos el ttulo

plt.show()#Mostramos la figura
```

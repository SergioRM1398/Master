### Ejercicio 3
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Genera un array 2D de números aleatorios enteros de tamaño (100, 5) y escribe una función que encuentre los índices de los elementos máximos de cada columna utilizando únicamente operaciones vectorizadas (sin bucles explícitos). En el caso de existir más de un valor máximo, se devolverá el índice del primero.
</div>

```python
np.random.seed(2) #Así siempre usamos los mismos números aleatorios (que lo hace mas facil de comprobar)

m1 = np.random.randint(0,501,(100,5)) #Creamos una matriz (100,5) de nnumeros aleatorios entre el 0 y el 500

def elem_max(matriz):    #Definimos la funcion para una matriz cualquiera
    filas=np.argmax(matriz, axis=0) #Nos devuelve el índice del primer valor máximo de cada fila ordenado
    columnas=np.arange(0,len(m1[1])) #Nos crea un array con las columnas
    indices= list(zip(filas,columnas))
    return print('Los índices de los valores máximos son:' , indices)
    #Juntamos las filas con sus respectivas columnas para obtener el índice (fila, columna)

elem_max(m1) #Comprobamos
```

### Ejercicio 1
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    Carga en la variable <strong>titanic</strong> el DataFrame titanic proveído por la librería Seaborn, y utilizando el método groupby, calcula la edad media de los pasajeros según el puerto de embarque. Almacena el resultado en la variable "mean_age". Asegúrate de que en tu resultado solo estás incluyendo la variable "age" (y los nombres de los puertos, por supuesto).
</div>

```python
titanic = sns.load_dataset("titanic") #Cargamos dataset

mean_age=titanic.groupby('embark_town')['age'].mean() #Calculamos la media de edades por puerto de embarque en una variable

print(mean_age) #Comprobamos la variable
```

<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    A continuación, usando el método <i>fillna</i> <strong>una sola vez</strong>, sustituye (inplace) los valores nulos del campo "age" del DataFrame <strong>titanic</strong> por los valores obtenidos en el apartado anterior (es decir, sustituye en el DataFrame <strong>titanic</strong> cada edad nula por el valor medio de la edad de los pasajeros que hayan embarcado en el mismo puerto).
</div>

```python
titanic = sns.load_dataset("titanic") #Cargamos dataset

for i in np.arange(0,len(mean_age)):
    #Creamos un filtro por comodidad usando los indices de mean_age
    filtro=(titanic['embark_town'] == mean_age.index[i]) & (titanic['age'].isna())
    #Aplicamos el filtro y cambiamos la variable NaN de age por su valor medio correspondiente dado por los indices de mean_age
    titanic.loc[filtro, 'age'] = titanic.loc[filtro, 'age'].fillna(mean_age[mean_age.index[i]])

print(titanic['age'].isnull().sum())
```

### Ejercicio 2
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Usando los datos de temperatura diaria de una ciudad indicados, calcula las medias móviles de 7 días y 30 días de la temperatura. Luego, identifica los días en los que la temperatura supera una desviación estándar por encima de la media móvil de 30 días.
</div>

```python
dates = pd.date_range(start='2022-01-01', end='2022-12-31')
np.random.seed(0)
temps = np.random.normal(loc = 20, scale = 5, size = len(dates))  # Temperaturas alrededor de 20 grados
```

```python
t=pd.DataFrame({ #Creamos el dataframe
    'fecha': dates,
    'temps': temps
})
t['media_7d']=t['temps'].rolling(window=7).mean() #Creamos una columna de promedios con la ventana movil de 7 días
t['media_30d']=t['temps'].rolling(window=30).mean() #Cramos una columna de promedios con la ventana movil de 30 días
print('Las fechas en las que la temperatura superó la media movil de 30 dias más su desviación estandar son:\n', t.loc[t['temps'] > t['media_30d'] + t['media_30d'].std(), 'fecha'])
#Mostramos los dias donde la temperatura media supero la media movil de 30 días + std
```

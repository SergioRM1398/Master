Genera un array 2D de nmeros aleatorios enteros de tamao (100, 5) y escribe una funcin que encuentre los ndices de los elementos mximos de cada columna utilizando nicamente operaciones vectorizadas (sin bucles explcitos). En el caso de existir ms de un valor mximo, se devolver el ndice del primero.
</div>

```python
np.random.seed(2) #As siempre usamos los mismos nmeros aleatorios (que lo hace mas facil de comprobar)

m1 = np.random.randint(0,501,(100,5)) #Creamos una matriz (100,5) de nnumeros aleatorios entre el 0 y el 500

def elem_max(matriz):    #Definimos la funcion para una matriz cualquiera
    filas=np.argmax(matriz, axis=0) #Nos devuelve el ndice del primer valor mximo de cada fila ordenado
    columnas=np.arange(0,len(m1[1])) #Nos crea un array con las columnas
    indices= list(zip(filas,columnas))
    return print('Los ndices de los valores mximos son:' , indices)
    #Juntamos las filas con sus respectivas columnas para obtener el ndice (fila, columna)

elem_max(m1) #Comprobamos
```

### Ejercicio 1
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    Carga en la variable <strong>titanic</strong> el DataFrame titanic provedo por la librera Seaborn, y utilizando el mtodo groupby, calcula la edad media de los pasajeros segn el puerto de embarque. Almacena el resultado en la variable "mean_age". Asegrate de que en tu resultado solo ests incluyendo la variable "age" (y los nombres de los puertos, por supuesto).
</div>

```python
titanic = sns.load_dataset("titanic") #Cargamos dataset

mean_age=titanic.groupby('embark_town')['age'].mean() #Calculamos la media de edades por puerto de embarque en una variable

print(mean_age) #Comprobamos la variable
```

<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
    A continuacin, usando el mtodo <i>fillna</i> <strong>una sola vez</strong>, sustituye (inplace) los valores nulos del campo "age" del DataFrame <strong>titanic</strong> por los valores obtenidos en el apartado anterior (es decir, sustituye en el DataFrame <strong>titanic</strong> cada edad nula por el valor medio de la edad de los pasajeros que hayan embarcado en el mismo puerto).
</div>

```python
titanic = sns.load_dataset("titanic") #Cargamos dataset

for i in np.arange(0,len(mean_age)):
    #Creamos un filtro por comodidad usando los indices de mean_age
    filtro=(titanic['embark_town'] == mean_age.index[i]) & (titanic['age'].isna())
    #Aplicamos el filtro y cambiamos la variable NaN de age por su valor medio correspondiente dado por los indices de mean_age
    titanic.loc[filtro, 'age'] = titanic.loc[filtro, 'age'].fillna(mean_age[mean_age.index[i]])

print(titanic['age'].isnull().sum())
```

### Ejercicio 2
<div style="background:#5D6D7E;padding:20px;color:#ffffff;margin-top:10px;">
Usando los datos de temperatura diaria de una ciudad indicados, calcula las medias mviles de 7 das y 30 das de la temperatura. Luego, identifica los das en los que la temperatura supera una desviacin estndar por encima de la media mvil de 30 das.
</div>

```python
dates = pd.date_range(start='2022-01-01', end='2022-12-31')
np.random.seed(0)
temps = np.random.normal(loc = 20, scale = 5, size = len(dates))  # Temperaturas alrededor de 20 grados
```

```python
t=pd.DataFrame({ #Creamos el dataframe
    'fecha': dates,
    'temps': temps
})
t['media_7d']=t['temps'].rolling(window=7).mean() #Creamos una columna de promedios con la ventana movil de 7 das
t['media_30d']=t['temps'].rolling(window=30).mean() #Cramos una columna de promedios con la ventana movil de 30 das
print('Las fechas en las que la temperatura super la media movil de 30 dias ms su desviacin estandar son:\n', t.loc[t['temps'] > t['media_30d'] + t['media_30d'].std(), 'fecha'])
#Mostramos los dias donde la temperatura media supero la media movil de 30 das + std
```

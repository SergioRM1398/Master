El proceso **Feature Selection** o **Selección de Variables** consiste esencialmente en elegir las variables relevantes comprobando si los datos de las mismas son relevantes o solo aportan ruido. Para ello se comprueba si la varianza de cada variable es **0** o cercana a él. En caso de que lo sea, implicaría que los datos tienen valores similares y podría añadir ruido, por lo que podría interesar eliminarlos.

El primer paso para realizar este proceso consiste en cargar el archivo **hatco.csv** y analizarlo.

```r
hatco <- read.csv('hatco.csv', sep= ';'); head(hatco) #Cargamos el csv e indicamos la separación

str(hatco[1:6,]) #Cargamos el string para visualizar la estructura

class(hatco) #Clase de la base de datos

dim(hatco) #Dimension de la base de datos

names(hatco) #Variables de la base de datos

nrow(hatco) #Comprobamos si hay o no datos faltantes
nrow(na.omit(hatco))
```

Vamos a analizarlos utilizando la función **`complete.cases()`** que nos devuelve un vector lógico, o de booleanos, que indica que filas están completas. Otra forma de analizarlo sería usando la función `na.omit()` ya que uno calcula las observaciones que están completas y la otra omite las observaciones que no lo esten.

```r
hatco[!complete.cases(hatco),] #Pedimos que nos muestre que filas están incompletas usando '!'
```

Como podemos observar solo la fila 100  tiene datos faltatantes. Como no aporta nada de información, podemos eliminarla del dataframe.

```r
hatco <- hatco[complete.cases(hatco),]; tail(hatco) #Hacemos que el dataframe sea aquel sin NA
dim(hatco) #La nueva dimensión del dataframe
```

Una vez hemos eliminado los datos irrelevantes del dataset, procedemos a interpretar los resultados:

El dataset **`hatco`** contiene datos sobre el porcentaje de compra de los clientes, recogido en la variable `fidelida`, así como el nivel de los precios y como estos varían, en las variables `precio`y `flexiprec`respectivamente. La calidad del producto viene definida por la variable `calidadp` y la imagen del fabricante y la fuerza que esta tiene en las ventas se encuentran definidas en `imgfabri`y en `imgvent`.

Una vez eliminada la fila que no aportaba información, se queda un dataset de $99$ observaciones sobre $6$ variables de tipo numérico. La función `class()` nos confirma que se trata de un dataframe, por lo que se encuentra clasificado en filas y columnas, con una dimensión de $(99,6)$.

Una vez realizada la presentación del conjunto de datos, podemos continuar con el procedimiento de **Feature Selection**.

Para calcular la varianza de cada una de las variables se utilizará la función **`nearZeroVar()`**, que se encuentra en la librería **caret**. El motivo de aplicación de esta función es que, si dejamos el argumento **`saveMetrics= TRUE`**, nos devolverá una columna en booleanos que indica si la varianza es cercana al valor **0** o no.

Tal y como se puede observar en el dataframe **varianza**, ninguna de las variables tiene una varianza igual o cercana a 0 (columnas **zeroVar** y **nzv** respectivamente). Esto implica que todas las variables aportan información y se pueden usar las variables numéricas.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

A continuación, escalamos los valores del dataframe usando la función **`preProcess()`** que se encuentra en la librería **caret** previamente importada.
Utilizaremos dos métodos de escalado:
+ Center: Centra la distribución en el origen de coordenadas restando los promedios de los valores para que den 0.
+ Range: normaliza los datos en un rango entre 0 y 1.

Comenzamos escalando con el método **center**:

```r
preprocess_center <- preProcess(hatco, method = 'center'); preprocess_center #No transforma los datos
hatco_center <- predict(preprocess_center, hatco); head(hatco_center) #Utilizamos la función predict() para mostrarlos
```

Seguimos ahora con **range**:

```r
preprocess_range <- preProcess(hatco, 'range'); preprocess_range
hatco_range <- predict(preprocess_range, hatco); head(hatco_range)
```

Una vez escalados los valores, analizamos si se han procesado correctamente los datos.

```r
cat("--- Datos sin procesar ---","\n",
    "Min:", min(hatco$fidelida),"\n",
    "Max:", max(hatco$fidelida), "\n","\n",
    "--- Datos procesados center ---","\n",
    "Min:", min(hatco_center$fidelida),"\n",
    "Max:", max(hatco_center$fidelida), "\n","\n",
    '--- Datos procesados range ---', '\n',
    "Min:", min(hatco_range$fidelida),"\n",
    "Max:", max(hatco_range$fidelida)
    )
```

Los valores se han acotado tal y como se ha indicado. Aquellos que se procesaron por el método center han acercado su valores al eje de coordenadas cartesianos reduciendo sus valores y los que se procesaron con el método de range se encuentran con valores entre 0 y 1.

Una vez comprobado el correcto procesado, se continúa eliminando  el 30% de los datos del dataframe.

```r
set.seed(13) #Creamos una semilla para reproducibilidad

n_elim = floor(0.30*(nrow(hatco)*ncol(hatco))); n_elim #Elegimos el número de variables a eliminar

indices = sample(1:(nrow(hatco)*ncol(hatco)), n_elim) #Creamos indices aleatorios
```

```r
# Función auxiliar para convertir índices lineales en índices de matriz
ind2ij = function(ind, nc) {
  cbind((ind - 1) %/% nc + 1, (ind - 1) %% nc + 1)
}

# Reemplazamos los valores seleccionados por NA en la variable hatcona
hatcona = hatco
hatcona[as.matrix(ind2ij(indices, ncol(hatcona)))] = NA

head(hatcona) #Mostramos las primeras filas
```

Para finalizar, imputamos los valores perdidos utilizando el método **`medianImpute`**

```r
pre_median <- caret::preProcess(hatcona, method = "medianImpute") #Imputamos los datos faltantes
imputed_median <- predict(pre_median, hatcona); head(imputed_median)
```

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

A continuación, usamos el método de `Análisis de Componentes Principales` o `PCA`, para reducir el número de variables manteniendo la mayor cantidad de información posible. Escogemos conservar un 80% de los datos, es decir, que perdemos un 20% de la información.

```r
pre_20 <- preProcess(hatco, 'pca', thresh = 0.8) #Eliminamos el 20% de la información

hatco_20 <- predict(pre_20, hatco); hatco_20 #Mostramos la tabla con el 80%
```

```r
pre_40 <- preProcess(hatco, 'pca', thresh = 0.6) #Eliminamos el 40% de la información
hatco_40 <- predict(pre_40, hatco); hatco_40 #Mostaramod la tabla con el 60%
```

Al asumir una pérdida del 20% de la información, la cantidad de columnas se reduce de 6 a 3, lo que indica una pérdida del 50% de los datos. Sin embargo, con una pérdida del 40%, las columnas resultantes son 2, es decir, se pierde un 66,67% de los datos.

De estos podemos decir que para el 40% de la pérdida de información perdemos tan solo un 16,67% más de datos que para una pérdida del 20%.

Para ello utilizaremos el **coeficiente de correlación de Pearson** evaluando ambas variables. Es importante recordar que el coeficiente de Pearson mide **relaciones lineales** entre variables numéricas y que la variable *`smoke`* es una variable booleana (1 o 0), es decir, categórica, aunque el coeficiente la tomará como una varibale numérica binaria (1 fumadoras y 0 no fumadoras).

Ahora, analicemos los posibles resultados del coeficiente de Pearson, que van de -1 a 1 donde:
* **1** implica una relación lineal positiva perfecta -> a mayor *`smoke`* (valor 1), mayor *`bwt`* (mayor peso).
* **-1** implica una relación lineal negativa perfecta -> a mayor *`smoke`* (valor 1), menor *`bwt`* (menor peso).
* **0** implica que NO existe una relación lineal entre las variables.

Luego, si el `p-valor` es menor que el nivel de significación $\alpha$, rechazamos la hipótesis nula $H_0$ y se concluye que hay una correlación significativa.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
test_pearson01 = cor.test(birthwt$smoke, birthwt$bwt, method= 'pearson', conf.level= 0.90) #Pearson para alfa igual a 0.1
print(test_pearson01)

test_pearson005 = cor.test(birthwt$smoke, birthwt$bwt, method= 'pearson', conf.level= 0.95) #Pearson para alfa igual a 0.05
print(test_pearson005)
```

Se calculó el `p-valor` de la hipótesis y se compara para los niveles de signifación $\alpha$=0.1 y $\alpha$=0.05. El `p-valor`obtenido es de 0.008667, ahora analizamos para cada caso:

**`p-valor` = 0.08667.** Por tanto:
* Para $\alpha$=0.1, `p-valor` < 0.1. Luego, rechazamos la hipótesis nula $H_0$ y concluimos que existe una correlación significativa.
* Para $\alpha$=0.05, `p-valor` < 0.05. Luego, volvemos a rechazar $H_0$ ya que la evidencia es aún más concluyente.

Podemos concluir que existe una evidencia estadística suficiente entre fumar durante el embarazo y  el peso de los bebes al nacer. Teniendo en cuenta que el signo del coeficiente de relación es negativo, la correlación también es negativa. Esto implica que es más probable que el bebé posea un peso más bajo si la madre fumó durante el embarazo.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Se ha calculado la correlación por medio del método de **Pearson**, que asume que los datos están **normalizados**. Es decir, que hemos realizado una prueba **paramétrica**. Sin embargo, es posible que los datos estén **no normalizados** y, por tanto, no habríamos realizado el método correcto ya que debería ser **no paramétrico**, como el de **Spearman** o el de **Kendall**.

Para comprobar si los datos están normalizados o no, los evaluaremos gráficamente y luego aplicaremos el **Test de Kolmogorov-Smirnov-Lilliefors (K-S-L)** donde, si rechazamos la hipótesis nula $H_0$ (`p-valor` $< \alpha$), los datos NO son normales.

```r
install.packages("nortest") #Instalamos el paquete nortest para usar el test de Lilliefors
```

```r
library(nortest) #Cargamos el paquete nortest
```

```r
hist(birthwt$smoke,freq=FALSE,col="gray",main='Histograma de la variable smoke', xlab= 'Fumadoras' ) #Histograma de la variable smoke
hist(birthwt$bwt,freq=FALSE,col="purple",main="Histograma de la variable bwt", xlab= 'Peso en gramos') #Histograma de la variable bwt (peso)

lillie_fum = lillie.test(birthwt$smoke); lillie_fum #Estudio de la normalización de la variable smoke
lillie_peso = lillie.test(birthwt$bwt); lillie_peso #Estudio de la normalización de la variable bwt (peso)
```

Observando los histogramas vemos como la variable referente al peso, `bwt`, se encuentra normalizada, a diferencia de la variable `smoke`. De la misma forma, al realizar el test de Lilliefors comprobamos que la varibale `smoke` rechaza la hipótesis nula $H_0$ por lo que podemos confirmar que no se encuentra normalizada. Por lo tanto, aunque para la variable `bwt` los datos sí estén parametrizados, es mejor usar un método no paramétrico.

A continuación, aplicaremos el método de **Spearman** y de **Kendall** a los datos para un $\alpha$=0.05.

```r
test_spearman = cor.test(birthwt$smoke, birthwt$bwt, method= 'spearman') #Metodo de Spearman
print(test_spearman)

test_kendall = cor.test(birthwt$smoke, birthwt$bwt, method= 'kendall') #Método de Kendall
print(test_kendall)
```

Tal y como observamos, para ambos casos obtenemos resultados parecidos al método de **Pearson**, por lo tanto, podemos llegar a la misma conclusión. Sin embargo, la correlación, en valor absoluto, es un poco menor para el método de **Kendall** que para los métodos de **Pearson** y **Spearman**, los cuales son bastante parecidos. Esto significa que la relación, según el método de **Kendall**, es más leve, comparada con los otros dos métodos. Lo cual es un resultado esperable ya que los valores por el método de **Kendall** suelen ser menores que por le método de **Spearman**.

```r
transacciones = c(100, -50, 0, 25) #Ponemos un vector para comprobar

rentability = function(transactions) {
    ren= c() #Creamos un vector vacio donde añadiremos los valores del vector inicial con las transacciones
    for (x in transactions) { #Hacemos un bucle que vaya leyendo cada valor del vector
        if (x > 0) { 
            ren = c(ren,'Ganancia') #Si el valor es positivo, que añada 'Ganancia' en esa posición
            }
        else if (x < 0) {
            ren = c(ren,'Pérdida') #Si es negativo, que añada 'Pérdida' en esa posición
            }
        else if (x == 0) {
            ren = c(ren, 'Neutral') #Y si es negativo, que sea 'Neutral'
            }
        }
    return (ren) #Que devuelva el vector creado con los datos rellenados
    }

rentability(transacciones)
```

Creamos la función `rentability()` que contenga un `bucle for` para que lea cada uno de los valores del vector a analizar y por medio de condicionales pedimos que añada el término Ganancia, Pérdida o Neutral al vector **`ren`** en función de si el valor tomado por el `bucle for` es mayor, menor o igual a 0.
